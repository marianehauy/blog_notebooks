```python
def mse(x):
    soma = 0
    for i in [36, 40, 42, 37, 38]:
        soma += np.power(i-x, 2)      
    return soma

x = np.linspace(33, 45, 200)
mse_ = list(map(mse, x))

x_min = x[np.argmin(mse_)]
mse_min = np.min(mse_)

fig, ax = plt.subplots(ncols=1, figsize=(12,6))
ax.set_title("MSE vs Valor Predito - K=5")
ax.set_xlabel("Valor Predito")
ax.set_ylabel("MSE")
ax.plot(x, mse_, lw=3, c="xkcd:sky blue")
ax.set_xticks([34,36,38,x_min,40,42,44])
ax.set_yticks([mse_min,50,100,150,200])
ax.set_ylim([0, 250])
ax.set_xlim([32, 46])
ax.axvline(x_min, 0, mse_min/250, c="grey", ls="dashed")
ax.axhline(mse_min, 0, (x_min-32)/(46-32), c="grey", ls="dashed")
fig.savefig("images/knn_5.png", dpi=300);
```

```python
def mink(p, x=(0,0), y=(2,2)):
    soma = 0
    for x_,y_ in zip(x, y):
        soma += np.power(np.abs(x_-y_), p)
    return np.power(soma, 1/p)

ps = [i for i in range(1,20)]

dists = list(map(mink, ps))

fig, ax = plt.subplots(figsize=(8,6))
ax.scatter(ps, dists, s=100, c="xkcd:sky blue")
ax.set_xlabel("Ordem (p)")
ax.set_ylabel("Distância")
ax.tick_params(bottom=False, left=False)
fig.savefig("images/knn_15.png", dpi=300);
```
