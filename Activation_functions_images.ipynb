```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def plot_function(x, func, func_name, color="xkcd:sky blue"):
    y = func(x)
    fig, ax = plt.subplots(figsize=(16,9))
    ax.plot(x, y, lw=5, c=color)
    ax.set_ylabel(f"{func_name}")
    ax.set_xlabel("x")
    ax.xaxis.label.set_fontsize(20)
    ax.xaxis.label.set_fontfamily("Sans Serif")
    ax.yaxis.label.set_fontsize(20)
    ax.yaxis.label.set_fontfamily("Sans Serif")
    return fig
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def deriv_sigmoid(x):
    return (sigmoid(x)*(1-sigmoid(x)))
```

```python
def clip(x):
    return np.clip(x, a_max=1, a_min=0)
```

```python
def deriv_clip(x):
    return np.where((x <= 1) & (x>=0), 1, 0)
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
```

```python
def deriv_tanh(x):
    return 1 - np.power(tanh(x), 2)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def deriv_relu(x):
    return np.where((x>0), 1, 0)
```

```python
x_lin = np.linspace(-10, 10, 500)
```

```python
fig = plot_function(x_lin, sigmoid, func_name="sigmoid(x)")
fig.savefig("images/func_ativ_01.png", dpi=150)
```

```python
x_lin = np.linspace(-1, 2, 500)
```

```python
fig = plot_function(x_lin, clip, func_name="clip(x, 0, 1)")
fig.savefig("images/func_ativ_02.png", dpi=150)
```

```python
fig = plot_function(x_lin, deriv_clip, func_name="clip'(x)", color="xkcd:salmon")
fig.savefig("images/func_ativ_03.png", dpi=150)
```

```python
x_lin = np.linspace(-10, 10, 500)
```

```python
fig = plot_function(x_lin, deriv_sigmoid, func_name="sigmoid'(x)", color="xkcd:salmon")
fig.savefig("images/func_ativ_04.png", dpi=150)
```

```python
fig = plot_function(x_lin, tanh, func_name="tanh(x)")
fig.savefig("images/func_ativ_05.png", dpi=150)
```

```python
fig, ax = plt.subplots(figsize=(16,9))
y = deriv_sigmoid(x_lin)
ax.plot(x_lin, y, lw=5, c="xkcd:salmon", label="$\sigma^{\prime}$")
y = deriv_tanh(x_lin)
ax.plot(x_lin, y, lw=5, c="xkcd:turquoise", label="$tanh^{\prime}$")
ax.set_ylabel(f"Derivada")
ax.set_xlabel("x")
ax.xaxis.label.set_fontsize(20)
ax.xaxis.label.set_fontfamily("Sans Serif")
ax.yaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontfamily("Sans Serif")
ax.legend(fontsize="x-large")
fig.savefig("images/func_ativ_06.png", dpi=150)
```

```python
fig = plot_function(x_lin, relu, func_name="relu(x)")
fig.savefig("images/func_ativ_07.png", dpi=150)
```

```python
fig = plot_function(x_lin, deriv_relu, func_name="relu'(x)", color="xkcd:salmon")
fig.savefig("images/func_ativ_08.png", dpi=150)
```

```python
fig, ax = plt.subplots(figsize=(16,9))
x_layers = list(range(1, 11))
y_relu = [deriv_relu(0.5)**i for i in range(1, 11)][::-1]
y_sigmoid = [deriv_sigmoid(0.5)**i for i in range(1, 11)][::-1]
y_tanh = [deriv_tanh(0.5)**i for i in range(1, 11)][::-1]
ax.plot(x_layers, y_sigmoid, lw=5, c="xkcd:sky blue", label="$\sigma$", ls="-.")
ax.plot(x_layers, y_tanh, lw=5, c="xkcd:salmon", label="$tanh$", ls="--")
ax.plot(x_layers, y_relu, lw=5, c="xkcd:turquoise", label="$ReLU$", ls=":")
ax.set_ylabel(f"$M$")
ax.set_xlabel("Camada")
ax.xaxis.label.set_fontsize(20)
ax.xaxis.label.set_fontfamily("Sans Serif")
ax.yaxis.label.set_fontsize(20)
ax.yaxis.label.set_fontfamily("Sans Serif")
ax.set_ylim(-0.05, 1.4)
ax.legend(fontsize="x-large")
fig.savefig("images/func_ativ_09.png", dpi=150)
```
